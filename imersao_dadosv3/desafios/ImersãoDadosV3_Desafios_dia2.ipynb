##Desafio 01: Deixar o gr√°fico countplot ordenado

###Decrescente

```python
import seaborn as sns
import matplotlib.pyplot as plt

cod_compostos = dados['composto'].value_counts().head().index

plt.figure(figsize=(10, 8))

data = dados.query('composto in @cod_compostos')

sns.set()
ax = sns.countplot(x='composto', data=data, order=cod_compostos)
ax.set_title('Top 5 compostos')

plt.show()
```

###Crescente

```python
cod_compostos_invertido = cod_compostos[::-1]

plt.figure(figsize=(10, 8))

ax_crescente = sns.countplot(x='composto', data=data, order=cod_compostos_invertido)
ax_crescente.set_title("Top 5 compostos")

plt.show()
```

##Desafio 02: Melhorar a visualiza√ß√£o do gr√°fico alterando o tamanho da fonte...

```python
plt.figure(figsize=(10, 8))

ax_estilizado = sns.countplot(x='composto', data=data, order=cod_compostos)
ax_estilizado.set_title('Top 5 compostos', fontsize=20, color="#5975A4")
ax_estilizado.set_xlabel('Composto', fontsize=20, color="#5975A4")
ax_estilizado.set_ylabel('Ocorr√™ncias', fontsize=20, color="#5975A4")

plt.show()
```

##Desafio 03: Plotar os histogramas com seaborn

```python
media_c = dados.loc[:, 'c0':'c99'].describe().T['mean']
media_c.head()
```

```python
plt.figure(figsize=(10, 8))

hist_c = sns.histplot(data=media_c, bins=50)
hist_c.set_title("Compostos C", fontsize=20, color='#5975A4')
hist_c.set_xlabel('M√©dia', fontsize=20, color='#5975A4')
hist_c.set_ylabel('Ocorr√™ncias', fontsize=20, color='#5975A4')

plt.show()
```
